```python
import numpy as np
import cv2
import os
```

```python
video_path = "./data/MOT16/train/MOT16-13"
os.chdir(video_path)
os.getcwd()
```

```python
video_name = video_path.split('/')[-1]
vidcap = cv2.VideoCapture(video_name+'.mp4')
success, image = vidcap.read()
os.chdir("./img")
count = 1
while success:
    img_name = str(count).zfill(4)
    print(img_name)
    cv2.imwrite("{}.jpg".format(img_name), image)     # save frame as JPEG file
    success, image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```
